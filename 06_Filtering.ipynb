# Filtering Data Exercise
Answer the following questions using the `bestsellers.csv` dataset

```python
import pandas as pd
bestsellers = pd.read_csv("data/bestsellers.csv")
bestsellers
```

## Find the books written by Pete Souza

### Solution

```python
bestsellers[bestsellers["Author"] == "Pete Souza"]
```

### Alternatively...

```python
pete_souza = bestsellers["Author"] == "Pete Souza"
bestsellers[pete_souza]
```

## Find the books that are under 10 dollars

### Solution

```python
bargains = bestsellers["Price"] < 10
bestsellers[bargains]
```

### Alternatively...

```python
bestsellers[bestsellers["Price"] < 10]
```

## Find the books that have a price between 50 and 60 dollars


### Solution

```python
bestsellers[bestsellers["Price"].between(50, 60)]
```

## Find all the books written by Kristin Hannah, Andy Weir, or Delia Owens


### Solution

```python
names = ["Kristin Hannah", "Andy Weir", "Delia Owens"]
bestsellers[bestsellers["Author"].isin(names)]
```

## Find the Non Fiction books that are rated 4.9


### Solution

```python
bestsellers[(bestsellers["Genre"] == "Non Fiction") & (bestsellers["User Rating"] == 4.9)]
```

### Alternatively, a little cleaner...

```python
nonfiction = bestsellers["Genre"] == "Non Fiction"
high_rating = bestsellers["User Rating"] == 4.9
bestsellers[nonfiction & high_rating]
```

## Find the fiction book with the lowest User Rating


### Solution - lol it's JK Rowling

```python
bestsellers[bestsellers["Genre"] == "Fiction"].sort_values("User Rating")
```

## Find 2012's top 5 Fiction books with the most Reviews


### Solution - Wasn't sure if it was asking to sort by rating AND # reviews, so I did both

* Just sorting by # reviews

```python
year_2012 = bestsellers["Year"] == 2012
fiction = bestsellers["Genre"] == "Fiction"
bestsellers[year_2012 & fiction].sort_values("Reviews", ascending = False).head()
```

* Sorting by rating and then review - I dont think this was what he was asking

```python
year_2012 = bestsellers["Year"] == 2012
fiction = bestsellers["Genre"] == "Fiction"
bestsellers[year_2012 & fiction].sort_values(["User Rating", "Reviews"], ascending = False).head()
```

## Create a bar plot showing the 5 authors who have the most books with a rating under 4.5


### Solution

```python
rating = bestsellers["User Rating"] < 4.5
bestsellers[rating]
```

```python
bestsellers[rating].value_counts("Author")
```

```python
bestsellers[rating].value_counts("Author").head().sort_values().plot(kind = "bar")
```
