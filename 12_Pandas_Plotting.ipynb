## Part 1

Create the bar plot seen in the image below.  It visualizes the top 10 artists who have spent the most weeks at number 1 in the dataset.

* Notice the customized title
* The figure is 8x6
* Add in the x-axis label
* The plot uses the ggplot matplotlib theme
* The bars have a black edge with a width of 3.  You'll need to add that in yourself!



### Solution

```python
import pandas as pd
import matplotlib.pyplot as plt

billboard = pd.read_csv("data/billboard_charts.csv", parse_dates = ["date"])
```

```python
no_1s = billboard[billboard["rank"] == 1]
no_1s
```

```python
top_10 = no_1s["artist"].value_counts().head(10)
top_10
```

```python
plt.style.available
```

```python
plt.style.use("ggplot")

top_10.plot(kind = "barh",
            figsize = (8,6),
            #x = "Weeks at #1",
            title = "Artists with Most Weeks at #1",
            edgecolor = "black",
            linewidth = 3
           )
plt.xlabel("Weeks at #1")
```

## Part 2

Please create the pie chart seen in the image below.  It visualizes the various artists who charted with the song "Please Mr. Postman".  

* Add in the custom title
* Use this list of colors: `colors = ["#6c5ce7","#00cec9", "#74b9ff"]`
* Explode the "Gentle Persuasion" slice by 0.1



### Solution

```python
postman = billboard[billboard["song"] == "Please Mr. Postman"]
postman
```

```python
postman["artist"].value_counts().plot(kind = "pie",
                                     figsize = (10,10),
                                     fontsize = 20,
                                     colors = ["#6c5ce7","#00cec9", "#74b9ff"],
                                     explode = (0, 0, 0.1)
                                     )
plt.title("Please Mr. Postman Versions",fontsize = 30)
plt.ylabel("artist", fontsize = 30)
```

## Part 3
Create the line plot seen in the image below. It visualizes the performance of the song "Blinding Lights"

* The figure is 9x5
* The line color is `#a29bfe`
* To invert the y-axis (so that rank 1 is the highest value), use: `plt.gca().invert_yaxis()`



### Solution

```python
blinding_lights = billboard[billboard["song"] == "Blinding Lights"].set_index("date")["rank"]
```

```python
plt.figure(figsize = (9, 5), dpi = 200)

blinding_lights.sort_index().plot(color = "#a29bfe", legend = None, linewidth = 2)

plt.yticks([0, 10, 20, 30, 40, 50, 60, 70])
plt.gca().invert_yaxis()
plt.title("Blinding Lights Chart Performance")
```

## Part 4

Create the line plot seen in the image below.  It shows the chart performance (rank) of the following songs:

* All I Want For Christmas Is You by Mariah Carey
* Rockin' Around The Christmas Tree by Brenda Lee
* Jingle Bell Rock by Bobby Helms

**The date range spans from 2016-12-25 to 2021-01-01**

Notice the customized x-axis tick marks, the legend, the title, and the axis labels! Also the figure is `10x7`

To invert the y-axis, use `plt.gca().invert_yaxis()`



### Solution - define df to have index of date and column "rank" so you don't need to specify x= and y= later

```python
all_i_want = billboard[(billboard["song"] == "All I Want For Christmas Is You") 
                       & (billboard["artist"] == "Mariah Carey") 
                       & (billboard["date"].between("2016-12-25", "2021-01-01"))].set_index("date")["rank"]
rockin = billboard[(billboard["song"] == "Rockin' Around The Christmas Tree") 
                   & (billboard["artist"] == "Brenda Lee") 
                   & (billboard["date"].between("2016-12-25", "2021-01-01"))].set_index("date")["rank"]
jingle = billboard[(billboard["song"] == "Jingle Bell Rock") 
                   & (billboard["artist"] == "Bobby Helms") 
                   & (billboard["date"].between("2016-12-25", "2021-01-01"))].set_index("date")["rank"]
```

```python
plt.figure(figsize = (10, 7))

all_i_want.plot()
rockin.plot()
jingle.plot()

plt.xticks(
    ["2016-12-25", "2017-12-25","2018-12-25","2019-12-25","2020-12-25"],
    labels=["Xmas 2016", "Xmas 2017","Xmas 2018","Xmas 2019","Xmas 2020",]
)
plt.title("Christmas Songs On The Hot 100")
plt.ylabel("Chart Rank")
plt.gca().invert_yaxis()
```

## Part 5

Create the visualization seen in the image below.  It shows the top 8 songs that spent the most time on the Billboard charts (at any position) for each of the 4 Beatle's solo acts:
* John Lennon
* Wings
* George Harrison
* Ringo Starr

The details:
* 2 rows, 2 columns
* Figure is 14x8
* The main title has a fontsize of 20
* Notice the colors of each plot and its corresponding title!
* Colors are: 
    * red: #fc5c65
    * orange: #fd9644
    * blue: #2bcbba
    * green: #26de81
    


### Solution

```python
plt.style.use("ggplot")
```

```python
john = billboard[billboard["artist"] == "John Lennon"]["song"].value_counts().head(8)
paul = billboard[billboard["artist"] == "Wings"]["song"].value_counts().head(8)
george = billboard[billboard["artist"] == "George Harrison"]["song"].value_counts().head(8)
ringo = billboard[billboard["artist"] == "Ringo Starr"]["song"].value_counts().head(8)
```

```python
fig, axs = plt.subplots(2, 2, figsize = (14,8))
fig.suptitle("Beatles Solo Songs (By Weeks on Chart)", fontsize = 20)


john.plot(kind = "barh", ax = axs[0][0], color = "#fc5c65")
axs[0][0].set_title(label = "John Lennon", color = "#fc5c65")

paul.plot(kind = "barh", ax = axs[0][1], color = "#fd9644")
axs[0][1].set_title(label = "Paul McCartney (Wings)", color = "#fd9644")

george.plot(kind = "barh", ax = axs[1][0], color = "#2bcbba")
axs[1][0].set_title(label = "George Harrison", color = "#2bcbba")

ringo.plot(kind = "barh", ax = axs[1][1], color = "#26de81")
axs[1][1].set_title(label = "Ringo Starr", color = "#26de81")

plt.tight_layout()
```

# UFO Plotting Challenge
## From video 134

## Solution - could also use .xlim to limit the years

```python
ufos = pd.read_csv("data/nuforc_reports.csv", parse_dates=['date_time', "posted"])
ufos
```

```python
ufos_2000 = ufos[ufos["date_time"].dt.year.between(2000, 2018)]

circle = ufos_2000[ufos_2000["shape"] == "circle"]
light = ufos_2000[ufos_2000["shape"] == "light"]
triangle = ufos_2000[ufos_2000["shape"] == "triangle"]
fireball = ufos_2000[ufos_2000["shape"] == "fireball"]
formation = ufos_2000[ufos_2000["shape"] == "formation"]

years = [2000, 2002, 2004, 2006, 2008, 2010, 2012, 2014, 2016, 2018]

#.value_counts().sort_index()
```

```python
plt.figure(figsize = (10, 8))

circle["date_time"].dt.year.value_counts().sort_index().plot(label = "circle")
light["date_time"].dt.year.value_counts().sort_index().plot(label = "light")
triangle["date_time"].dt.year.value_counts().sort_index().plot(label = "triangle")
fireball["date_time"].dt.year.value_counts().sort_index().plot(label = "fireball")
formation["date_time"].dt.year.value_counts().sort_index().plot(label = "formation")

plt.xticks(years)
plt.title("UFO Sightings by Shape")
plt.legend()
```
